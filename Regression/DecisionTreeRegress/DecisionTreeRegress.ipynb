# Decision Tree Regression
very useful for multi dimensional dataset as compared to single dimensional


Import Library


```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:, 1:-1].values
y= df.iloc[:,-1].values
```

for decision tree and random forest not to use feature scaling 

training decision tree on whole dataset


```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(random_state= 0)
reg.fit(X,y)
```

```python
print(reg.predict([[6.5]]))
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X))
plt.title("Salaries vs position")
plt.xlabel("position")
plt.ylabel('salaries')
plt.show()
```

```python
#high resolution
X_grid = np.arange(min(X),max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid) , 1))
plt.scatter(X,y, color='red')
plt.plot(X_grid, reg.predict(X_grid), color='blue')
plt.show()
```

```python
from sklearn.metrics import r2_score
# Assuming `X_test` is your test data and `y_test` are the true values
predicted_values = reg.predict(X)
r2_score(y, predicted_values)
```
