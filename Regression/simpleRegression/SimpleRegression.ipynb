The Ordinary Least Squares (OLS) method is a statistical technique used in regression analysis to estimate the parameters of a linear regression model. In simple terms, it is a method to find the best-fitting line through a set of data points.

Here's a brief overview of how OLS works:

1. **Model Representation:** In linear regression, the relationship between the independent variable(s) and the dependent variable is represented by a linear equation. For a simple linear regression with one independent variable, the equation is of the form:

   \(y = b_0 + b_1x + \varepsilon\)

   where:
   - \(y\) is the dependent variable.
   - \(x\) is the independent variable.
   - \(b_0\) is the y-intercept.
   - \(b_1\) is the slope of the line.
   - \(\varepsilon\) is the error term.

2. **Minimizing the Residuals:** OLS aims to minimize the sum of the squared differences between the observed values of the dependent variable (\(y\)) and the values predicted by the linear equation. These differences are called residuals. Mathematically, the objective is to minimize the sum of squared residuals:

   \[ \sum_{i=1}^{n} (y_i - (b_0 + b_1x_i))^2 \]

   where \(n\) is the number of data points.

3. **Deriving the Estimators:** By taking the partial derivatives of the above expression with respect to \(b_0\) and \(b_1\), and setting them to zero, we can derive the estimators (coefficients) that minimize the sum of squared residuals.

4. **Coefficients Calculation:** The derived estimators are used to calculate the values of \(b_0\) and \(b_1\) that best fit the data.

OLS is widely used due to its simplicity and efficiency, especially when the assumptions of the linear regression model are met. These assumptions include linearity, independence of errors, homoscedasticity (constant variance of errors), and normality of errors. If these assumptions are violated, it might affect the reliability of the OLS estimates.

importing the libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

Importing the dataset


```python
dataset = pd.read_csv("Salary_Data.csv")
X = dataset.iloc[: , :-1].values
y = dataset.iloc[: ,-1].values
```

Splitting the dataset  into training and  test


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Training the simple Linear Regression model on the training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Predicting the test result

```python
y_predicted = regressor.predict(X_test)
```

Visualising the training set result

```python
plt.scatter(X_train, y_train , color='red')
plt.plot(X_train, regressor.predict(X_train))
plt.title("Simple Linear Regression of Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, regressor.predict(X_test))
plt.title("Test Result")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

Making a single prediction (for example the salary of an employee with 12 years of experience)

```python
print(regressor.predict([[12]]))
```

Getting the coefficients and intercept

```python
print(regressor.coef_)
print(regressor.intercept_)
```
